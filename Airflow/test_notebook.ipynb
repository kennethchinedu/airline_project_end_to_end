```python
# Save new file on every load

import requests
import pandas as pd
from dotenv import load_dotenv
import os
import json
from datetime import datetime

start_year = '2024'
end_year = '2024'
order_by = 'date'
limit = '100'


def extract_titles():
    # Construct the current timestamp for the CSV file name
    timestamp = datetime.now().strftime('%Y%m%d%H%M%S')

    # Construct the CSV file name with the timestamp
    csv_file_name = f'Netflix2024_{timestamp}.csv'

    # Construct URL with start_year, end_year, order_by
    url = f"https://api.apilayer.com/unogs/search/titles?start_year={start_year}&order_by={order_by}&end_year={end_year}&limit={limit}"
    
    payload = {}
    headers = {
        "apikey": "LQpuQgeu4U6qvBAKZd4TcJeDHGDh4RFf"
    }

    response = requests.request("GET", url, headers=headers, data=payload)
    data = response.json()

    # Extract relevant fields from each result
    formatted_data = []
    for result in data["results"]:
        formatted_data.append({
            "imdb_id": result["imdb_id"],
            "title": result["title"],
            "rating": result["rating"],
            "year": result["year"],
            "runtime": result["runtime"],
            "top250": result["top250"],
            "top250tv": result["top250tv"],
            "title_date": result["title_date"]
        })

    # Create a DataFrame from the formatted data
    new_df = pd.DataFrame(formatted_data)

    # Save the new DataFrame as a new CSV file
    new_df.to_csv(csv_file_name, index=False)

    print(f"CSV file {csv_file_name} saved successfully.")
    # Display the DataFrame
    return new_df
```

```python
## Append


import requests
import pandas as pd
from dotenv import load_dotenv
import os, json

start_year = '2024'
end_year = '2024'
order_by = 'date'
limit = '20'

def extract_titles22():
    # Check if the CSV file exists
    if os.path.exists('Netflix2024.csv'):
        # Load existing data from CSV
        existing_df = pd.read_csv('Netflix2024.csv')
        # Get the latest title_date from the existing data
        latest_title_date = existing_df['title_date'].max()
    else:
        # If CSV file doesn't exist, set latest_title_date to None
        latest_title_date = None

    # Construct URL with start_year, end_year, order_by, and latest_title_date
    url = f"https://api.apilayer.com/unogs/search/titles?start_year={start_year}&order_by={order_by}&end_year={end_year}&limit={limit}"
    if latest_title_date:
        url += f"&title_date_from={latest_title_date}"
    
    payload = {}
    headers= {
        "apikey": "LQpuQgeu4U6qvBAKZd4TcJeDHGDh4RFf"
    }

    response = requests.request("GET", url, headers=headers, data = payload)
    data = response.json()

    # Extract relevant fields from each result
    formatted_data = []
    for result in data["results"]:
        formatted_data.append({
            "imdb_id": result["imdb_id"],
            "title": result["title"],
            "rating": result["rating"],
            "year": result["year"],
            "runtime": result["runtime"],
            "top250": result["top250"],
            "top250tv": result["top250tv"],
            "title_date": result["title_date"]
        })

    # Create a DataFrame from the formatted data
    new_df = pd.DataFrame(formatted_data)
    
    # Append new data to existing data (if any)
    if 'existing_df' in locals():
        combined_df = pd.concat([existing_df, new_df], ignore_index=True)
    else:
        combined_df = new_df
    
    # Save the combined DataFrame as a CSV file
    combined_df.to_csv('Netflix2024.csv', index=False)

    print("CSV file saved successfully.")
    # Display the DataFrame
    return combined_df
```
